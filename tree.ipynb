```python
import os

def generate_tree(root_dir, allowed_extensions=('.py', '.yaml', '.yml', '.md', 'requirements.txt'), indent=' '):
    """
    Generate a tree structure of the directory including only specific file types.

    Parameters:
    - root_dir: The root directory to generate the tree from.
    - allowed_extensions: A tuple of file extensions to include in the tree.
    - indent: The indentation to use for the tree structure.
    """
    for root, dirs, files in os.walk(root_dir):
        level = root.replace(root_dir, '').count(os.sep)
        indent_level = ' ' * 4 * (level)
        print('{}{}/'.format(indent_level, os.path.basename(root)))
        sub_indent = ' ' * 4 * (level + 1)
        for f in files:
            if f.endswith(allowed_extensions) or f == 'requirements.txt':
                print('{}{}'.format(sub_indent, f))

# Example usage
root_directory = '.'
print(root_directory + "/")
generate_tree(root_directory)
```
